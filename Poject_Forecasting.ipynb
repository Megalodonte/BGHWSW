```python
from statsmodels.tsa.arima_model import ARIMA
import pmdarima as pm
import pandas as pd

df = pd.read_csv('February_2022.csv', sep=',')
df
```

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df
```

```python
y=df['Melbourne Central']
a=df['Date']

```

```python
y=y.to_frame()
```

```python
y['index']=a
y=y.set_index('index')
y
```

```python
y=y.groupby('index').sum()
y
```

```python
import matplotlib.pyplot as plt
y.plot(title='count')
plt.show()
```

```python
test=pd.read_csv('March_2022.csv')
test['Date'] = pd.to_datetime(test['Date'], dayfirst=True)
t=test['Melbourne Central']
b=test['Date']
t=t.to_frame()
t['index']=b
t=t.set_index('index')
t=t.groupby('index').sum()


```

```python
model = pm.auto_arima(y['Melbourne Central'], start_p=1, start_q=1,
                      test='adf',       # use adftest to find optimal 'd'
                      max_p=3, max_q=3, # maximum p and q
                      m=1,              # frequency of series
                      d=None,           # let model determine 'd'
                      seasonal=False,   # No Seasonality
                      start_P=0, 
                      D=0, 
                      trace=True,
                      error_action='ignore',  
                      suppress_warnings=True, 
                      stepwise=True)

print(model.summary())
```

```python
from statsmodels.tsa.stattools import adfuller
from numpy import log
import numpy as np
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
```

```python
model.plot_diagnostics(figsize=(7,5))
plt.show()
```

```python
# Forecast
n_periods = len(t)
fc, confint = model.predict(n_periods=n_periods, return_conf_int=True)


index_of_fc = t.index

# make series for plotting purpose
fc_series = pd.Series(fc, index=index_of_fc)
lower_series = pd.Series(confint[:, 0], index=index_of_fc)
upper_series = pd.Series(confint[:, 1], index=index_of_fc)

# Plot
plt.plot(y['Melbourne Central'])
plt.plot(t["Melbourne Central"], color="orange")
plt.plot(fc_series, color='darkgreen')
plt.fill_between(lower_series.index, 
                 lower_series, 
                 upper_series, 
                 color='k', alpha=.15)

plt.title("Final Forecast")
plt.show()
```
